# 1. Tìm hiểu các loại ung thư phổi

- Để có thể xây dựng được một mô hình nhận dạng ung thư phổi thông qua ảnh chụp CT-Scan ở ngực thì với vị trí là đại diễn kĩ thuật ta rất cần hiểu sơ lược về kiến thức y khoa đối với loại bệnh này. > 1. Adenocarcinoma – Ung thư biểu mô tuyến

> 3. Squamous Cell Carcinoma – Ung thư biểu mô tế bào vảy
Chiếm khoảng 30% trong tổng số các trường hợp ung thư phổi không phải tế bào nhỏ (NSCLC). Thường liên quan đến việc hút thuốc lá và được tìm thấy ở trung tâm của phổi, nơi các phế quản lớn nối với khí quản và phổi.

## 3.1. TensorFlow

- TensorFlow là một thư viện mã nguồn mở của Google, được sử dụng rộng rãi để xây dựng và huấn luyện các mô hình học sâu. Một số điểm nổi bật của - TensorFlow bao gồm:

> - Đa nền tảng: TensorFlow có thể chạy trên nhiều nền tảng khác nhau, bao gồm CPU, GPU và TPU, giúp tối ưu hóa hiệu suất xử lý.
> - Thao tác tensor: TensorFlow làm việc chủ yếu với các tensors, là các mảng dữ liệu đa chiều. Điều này giúp xử lý các phép toán đại số tuyến tính phức tạp một cách hiệu quả.
> - Thư viện phong phú: TensorFlow cung cấp nhiều API cấp cao như Keras để dễ dàng xây dựng và huấn luyện các mô hình học sâu.
> - Cộng đồng và tài liệu: Với một cộng đồng người dùng lớn và nhiều tài liệu hỗ trợ, TensorFlow rất thân thiện với người mới bắt đầu cũng như những chuyên gia.

## 3.2. Mô hình Nơ-ron Tích chập (CNN)

- Mô hình nơ-ron tích chập (CNN) là một loại mạng nơ-ron nhân tạo đặc biệt mạnh mẽ trong các bài toán về xử lý ảnh và nhận diện đối tượng. CNN có một số đặc điểm chính sau:
> - Lớp tích chập (Convolutional Layer): Đây là lớp cốt lõi của CNN, nơi các bộ lọc (filters) hoặc các hạt nhân (kernels) được sử dụng để phát hiện các đặc trưng cục bộ từ dữ liệu đầu vào. Bộ lọc này trượt qua ảnh và thực hiện phép tích chập để tạo ra bản đồ đặc trưng (feature map).
> - Lớp gộp (Pooling Layer): Sau lớp tích chập, lớp gộp thường được sử dụng để giảm kích thước của bản đồ đặc trưng và giảm thiểu số lượng tham số và tính toán. Lớp gộp phổ biến là gộp cực đại (max pooling).
> - Lớp kết nối đầy đủ (Fully Connected Layer): Các lớp cuối cùng trong CNN thường là các lớp kết nối đầy đủ, nơi các neuron kết nối với tất cả các neuron trong lớp trước đó. Các lớp này giúp kết hợp các đặc trưng đã trích xuất để đưa ra quyết định cuối cùng (ví dụ: phân loại ảnh).
> - Tính không gian: CNN khai thác tính không gian của dữ liệu, tức là các mối quan hệ cục bộ giữa các điểm dữ liệu (như các điểm ảnh trong một hình ảnh), làm cho nó đặc biệt hiệu quả trong các bài toán liên quan đến thị giác máy tính.

- Ứng dụng
> - Xử lý ảnh: Nhận diện đối tượng, phân loại ảnh, phát hiện biên, v.v.
> - Thị giác máy tính: Nhận diện khuôn mặt, phân tích video, tự lái xe, v.v.
> - Xử lý ngôn ngữ tự nhiên: Mặc dù ít phổ biến hơn, CNN cũng được áp dụng trong các bài toán như phân loại văn bản và phân tích cảm xúc.

- Kiểm tra sơ đồ thư mục Data

```python
def print_directory_tree(root_dir, indent=""):
    try:
        items = os.listdir(root_dir)
    except PermissionError:
        print(indent + "Permission Denied")
        return
    
    directories = []
    file_count = 0
    
    for item in items:
        item_path = os.path.join(root_dir, item)
        if os.path.isdir(item_path):
            directories.append(item)
        else:
            file_count += 1
    
    # In thông tin các thư mục trước
    for directory in directories:
        print(indent + "|--> " + directory)
        print_directory_tree(os.path.join(root_dir, directory), indent + "    ")
    
    if file_count > 0:
        print(indent + "|--> " + f"{file_count} images")

directory_path = 'Data'
print_directory_tree(directory_path)
```

```python
train_valid_types = ['adenocarcinoma_left.lower.lobe_T2_N0_M0_Ib','large.cell.carcinoma_left.hilum_T2_N2_M0_IIIa','normal','squamous.cell.carcinoma_left.hilum_T1_N2_M0_IIIa']
test_types = ['adenocarcinoma','large.cell.carcinoma','normal','squamous.cell.carcinoma']
```

```python
number_data = []
fig, axs = plt.subplots(1, 3, figsize=(20, 15))
count = 0
for path in ['train','valid','test']:
    number_type = []
    if path != 'test':
        types = train_valid_types
    else:
        types = test_types
    for type_cancer in types:
        directory = f'Data/{path}/{type_cancer}'
        files = [file for file in os.listdir(directory) if os.path.isfile(os.path.join(directory, file))]
        number_type.append(len(files))
    number_data.append(sum(number_type))
    # Vẽ biểu đồ tròn
    types_show = [(t[0:15] + '...') if len(t) > 15 else t for t in types]
    axs[count].pie(number_type, labels=types_show, autopct='%1.1f%%', startangle=140)
    axs[count].set_title(f'Tỉ lệ các loại ung thư trong tập {path}')
    count+=1
plt.show()
```

```python
plt.pie(number_data, labels=['train','valid','test'], autopct='%1.1f%%', startangle=100)
plt.axis('equal')
plt.title('Tỉ lệ ảnh huấn luyện')
plt.show()
```

```python
# Các thư mục chứa ảnh
test_path = 'Data/test'
train_path = 'Data/train'
valid_path = 'Data/valid'

# Các thông số tiền xử lý ảnh load
IMG_SIZE = 224 # Resize ảnh về kích thước 256 x 256
BATCH_SIZE = 50 # Số lượng ảnh đưa vào mô hình mỗi một huấn luyện 

# Các thông số để load dữ liệu ảnh đối với train, test, valid
test_data_generate = tf.keras.preprocessing.image.ImageDataGenerator(
    dtype='float32',
    preprocessing_function=preprocess_input # tiền xử lý theo đầu vào của mô hình ResNet50
)

train_data_generate = tf.keras.preprocessing.image.ImageDataGenerator(
    dtype='float32',
    preprocessing_function=preprocess_input, # tiền xử lý theo đầu vào của mô hình ResNet50
    rotation_range=10,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    vertical_flip=False
)

valid_data_generate = tf.keras.preprocessing.image.ImageDataGenerator(
    dtype='float32',
    preprocessing_function=preprocess_input # tiền xử lý theo đầu vào của mô hình ResNet50
)
```

```python
test_data = test_data_generate.flow_from_directory(
    test_path,
    target_size= (IMG_SIZE,IMG_SIZE),
    batch_size= BATCH_SIZE,
    class_mode='categorical',
    shuffle=False
)

train_data = train_data_generate.flow_from_directory(
    train_path,
    target_size= (IMG_SIZE,IMG_SIZE),
    batch_size= BATCH_SIZE,
    class_mode='categorical'
)

valid_data = valid_data_generate.flow_from_directory(
    valid_path,
    target_size= (IMG_SIZE,IMG_SIZE),
    batch_size= BATCH_SIZE,
    class_mode='categorical'
)
```

## 4.4. Một số setup chuẩn bị cho mô hình

- Thông số mô hình

```python
CHANNELS = 3
EPOCH = 50
INPUT_SHAPE = (IMG_SIZE, IMG_SIZE, CHANNELS)
NUM_CLASSES = 4
```

- Setup Early Stopping để giám sát hiệu suất của mô hình trên một tập dữ liệu validation trong suốt quá trình huấn luyện. Khi hiệu suất trên tập validation không cải thiện trong một số lượng epoch nhất định (được gọi là patience), quá trình huấn luyện sẽ dừng lại để tránh overfitting

```python
earlystop = tf.keras.callbacks.EarlyStopping(monitor='val_accuracy',
                         min_delta=0.001,  # Điều chỉnh nếu cần
                         patience=15, # Số lượng Epoch nếu hiệu suất không cải thiện sẽ dừng lại
                         restore_best_weights=True)
```

- Setup Learning Rate Reducing Rate để tốc độ của learning rate giảm dần trong quá trình huấn luyện mạng neural. Khi huấn luyện mạng neural, việc điều chỉnh learning rate là một phần quan trọng để đảm bảo rằng quá trình huấn luyện diễn ra hiệu quả và mô hình hội tụ đến điểm tối ưu.

```python
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss',
                              factor=0.1,
                              patience=10,
                              verbose=1,
                              min_delta=0.0001,
                              min_lr=0.0001) 
```

```python
callbacks = [reduce_lr,earlystop]
```

- Điều chỉnh ResNet50 model

```python
ResNet50_model = tf.keras.applications.ResNet50(weights = 'imagenet',
                     include_top = False,
                     input_shape = INPUT_SHAPE)

for layer in ResNet50_model.layers:
    layer.trainable = False
```

## 4.5. Xây dựng mô hình

```python
model = models.Sequential([
    ResNet50_model,
    layers.BatchNormalization(),
    layers.MaxPooling2D(pool_size=(2, 2)),
    layers.Dropout(0.3),  
    layers.Flatten(),
    layers.Dense(1024, activation='relu'),
    layers.Dropout(0.3),  
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.3),  
    layers.Dense(256, activation='relu'),
    layers.Dropout(0.3),  
    layers.Dense(NUM_CLASSES, activation='softmax'),
])

model.build(input_shape=INPUT_SHAPE)
```

```python
model.summary()
```

```python
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), 
    loss='categorical_crossentropy', 
    metrics=['accuracy']
)
```

```python
history = model.fit(
    train_data,
    epochs=EPOCH,
    batch_size=BATCH_SIZE,
    verbose=1,
    validation_data=valid_data,
    callbacks=callbacks
)
```

## 4.6. Đánh giá mô hình

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']
```

```python
plt.figure(figsize=(8,8))

plt.subplot(1, 2, 1)
plt.plot(range(EPOCH), acc, label='Training Accuracy')
plt.plot(range(EPOCH), val_acc, label='Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(range(EPOCH), loss, label='Training Loss')
plt.plot(range(EPOCH), val_loss, label='Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')


plt.show()
```

- Nhận xét: Mô hình đã được huấn luyện hiệu quả song ở những Epoch cuối mô hình lại có dấu hiệu overfitting nhưng chung quy lại với mục tiêu tìm hiểu và xây dựng mô hình nhận dạng ung thư phổi ta đã thành công trong việc ứng dụng mô hình CNN cũng như giải quyết được các vấn đề trong việc thử nghiệm

```python
num_test_samples = len(test_data)
num_classes = len(test_data.class_indices)

predicted_probabilities = model.predict(test_data, steps=num_test_samples)
predicted_labels = np.argmax(predicted_probabilities, axis=1)

true_labels = test_data.classes

report = classification_report(true_labels, predicted_labels)

print(report)

test_c = list(test_data.class_indices.keys())
cm = confusion_matrix(true_labels, predicted_labels)
print("\n\nConfusion Matrix:\n", cm)

plt.figure(figsize=(10,4))
sns.heatmap(cm, annot=True, fmt='g', cmap='Greens', xticklabels = test_c, yticklabels = test_c)
plt.xlabel('\n\nPredicted Label\n')
plt.ylabel('\nTrue Label\n')
plt.title('Confusion Matrix\n\n')
plt.show()
```
